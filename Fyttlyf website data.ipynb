```python
def payload_split():
    payload_key=[]
    payload_value=[]
    for i in testing['payload']:
        sp=i.split("&")
        payload_key.append(sp[0])
        payload_value.append(sp[1])
    testing['payload_key']=payload_key
    testing['payload_value']=payload_value
    testing.drop('payload',axis=1,inplace=True)
    testing.to_csv('payload_split')
```
